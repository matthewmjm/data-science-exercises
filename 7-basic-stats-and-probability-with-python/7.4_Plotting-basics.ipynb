```python
import numpy as np
import pandas as pd
# Make the random function consistent and replicable
np.random.seed(1228)

# Make a blank data frame
df = pd.DataFrame()

# Add a column of random numbers between 0 and 1
df['rand'] = np.random.rand(100)
df['rand_sq'] = df['rand'] ** 2
df['rand_shift'] = df['rand'] + 2

# When creating a data frame an index column of counts is created, counting from zero.
# Here we do a few transforms on that index to create some extra columns.
df['counts_sq'] = df.index ** 2
df['counts_sqrt'] = np.sqrt(df.index)
```

```python
df.head(10)
```

```python
%matplotlib inline
plt.plot(df['rand'])
plt.show()
```

```python
%matplotlib inline
plt.plot(df['rand_sq'])
plt.show()
```

```python
%matplotlib inline
plt.plot(df['rand_shift'])
plt.show()
```

```python
%matplotlib inline
plt.plot(df['counts_sq'])
plt.show()
```

```python
%matplotlib inline
plt.plot(df['counts_sqrt'])
plt.show()
```

```python
# BASIC OPTIONS
```

```python
%matplotlib inline
plt.plot(df['rand'], color='purple')
plt.ylim([-0.1, 1.1])
plt.ylabel('VALUES')
plt.xlabel('RANDOM SERIES')
plt.show()
```

```python
%matplotlib inline
plt.plot(df['rand'], color='purple')
plt.plot(df['rand_shift'], color='green')
plt.ylim([-0.1, 3.1])
plt.ylabel('VALUES')
plt.xlabel('RANDOM SERIES')
plt.show()
```

```python
# SCATTER
```
