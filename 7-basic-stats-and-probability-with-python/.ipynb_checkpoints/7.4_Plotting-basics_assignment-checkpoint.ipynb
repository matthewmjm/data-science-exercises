```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('./files/iris.csv')
print(df)
```

```python
# line plot
plt.plot( df["sepal.length"])
plt.show()
```

```python
## This line plot shows sepal length for all 150 Iris in the data set.  The length
## shows along the Y-axis which looks like it falls between 4.3-7.8.
```

```python
# scatterplot
plt.scatter(x=df["sepal.length"], y=df["sepal.width"], color='green')
plt.scatter(x=df["petal.length"], y=df["petal.width"], color='red' )
plt.show()
```

```python
## This scatter plot shows the relationship between sepal and petal size for 
## all 150 Iris in the data set.  The length of the sepal/petal is along the X-axis and the
## sepal/petal width is shown along the Y-axis (between 0-4)
```

```python
# histogram
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.hist(x=df["sepal.width"])
plt.show()
```

```python
## This historgram shows the sepal width of all 150 Iris in the dataset.  The majority
## fall in the 3.0-3.5 width (center) and the standard deviation is .5.
```

```python
# boxplot
plt.boxplot(x=df["petal.width"])
plt.show()
```

```python
## This boxplot shows the petal width of all 150 Iris flowers in the dataset.  The "box"
## shows the interquartile range between .3-1.8.  Therefore that is where the 25th percentile
## to the 75th percentile fall.  The median is represented by the line, at about 1.4.
## The whiskers above and below the box represent the 0-25 percentile and 75-100 percentile.
## which puts all the petal width between .1 through 2.5 in petal width.
```
